## Visualization

```python
# visualizing the data through bar
def plot_bar(x,y,title,xlabel="x",ylabel="y"):
    plt.bar(x,y)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```

```python
# number of males vs females
count = data["Sex"].value_counts()
x = ["Male","Female"]
y = [count.iloc[0],count.iloc[1]]
plot_bar(x,y,"Male vs Female","Category","Count")
```

```python
# countries with less heart attack
def country_and_attack(title,hasAttack = 0):
    new_data = data[["Country","Heart Attack Risk"]]
    no_risk = new_data[new_data["Heart Attack Risk"] == hasAttack] # selecting only rows that 0 in heart attack risk
    count_country = no_risk["Country"].value_counts() # counting values for each countries
    top_countries = count_country.nlargest(5) # selecting only top 5

    plt.figure(figsize=(10,6))
    plt.bar(top_countries.index,top_countries.values)
    plt.xlabel("Countries")
    plt.ylabel("Count")
    plt.title(title)
    plt.show()
    
country_and_attack("Least Heart Attack",hasAttack = 0)
```

```python
country_and_attack("Most Attacks",hasAttack=1)
```
